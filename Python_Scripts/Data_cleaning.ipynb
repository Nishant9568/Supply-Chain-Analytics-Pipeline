```python
import pandas as pd
file_path = 'data/DataCoSupplyChainDataset.csv'
print("Data loading please wait...")
df = pd.read_csv(file_path, encoding='latin-1')
print("\nData loaded successfully! ✅")
print("Total Rows & Columns:", df.shape)
pd.set_option('display.max_columns', None) 
df.head()
```

```python
# 1. Jo columns humein analysis aur dashboard ke liye chahiye
columns_needed = [
    'Type', 'Days for shipping (real)', 'Days for shipment (scheduled)', 
    'Late_delivery_risk', 'Category Name', 'Customer City', 'Customer Segment', 
    'Order City', 'Order Country', 'order date (DateOrders)', 
    'shipping date (DateOrders)', 'Order Item Total', 'Order Status', 'Shipping Mode'
]

# 2. Naya dataframe banayein jisme sirf ye columns hon
df_clean = df[columns_needed].copy()

print("Columns filter ho gaye!")
print("Naye Dataset ka size:", df_clean.shape)
```

```python
df_clean.columns = [
    col.lower().replace(' ', '_').replace('(', '').replace(')', '')
    for col in df_clean.columns
]
print("Sql Friendly columns selected ")
print(df_clean.columns.tolist())
```

```python
print(df_clean.dtypes)
```

```python
df_clean['order_date_dateorders'] = pd.to_datetime(df_clean['order_date_dateorders'])
df_clean['shipping_date_dateorders'] = pd.to_datetime(df_clean['shipping_date_dateorders'])
print("Dates successfully converted to datetime format! 📅")
```

```python
print(df_clean.dtypes)
```

```python
import numpy as np 
df_clean['delay_in_days'] = df_clean['days_for_shipping_real'] - df_clean['days_for_shipment_scheduled']
df_clean['delivery_status'] = np.where(df_clean['delay_in_days'] > 0, 'Late', 'On-Time')

print("check new column /n:")
```

```python
df_clean.head()
```

```python
print("Check Missing Values")
df_clean.isnull().sum()
```

```python
print("Check Duplicates values")
duplicates = df_clean.duplicated().sum()
print(duplicates)
```

```python
df_clean = df_clean.drop_duplicates()
```

```python
# 3. Cleaned data ko 'Data' folder mein save karna
final_file_path = 'Data/final_cleaned_supply_chain.csv'
df_clean.to_csv(final_file_path, index=False)

print(f"\n✅ Congratulations! Cleaned dataset successfully saved at: {final_file_path}")
```
